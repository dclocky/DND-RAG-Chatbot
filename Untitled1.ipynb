```python
from dataclasses import dataclass
import re

@dataclass
class ExtractfromQoute:
  Supplier : str
  item : str
  unit_price : float
  delivery_time : str
  risk_notes : str

class ExtractorAgent:
  def __init__(self, LLM):
    self.model = SentenceTransformer()

  def Extract(self, text:str)->ExtractfromQoute:
    supplier = self.ExtractSupplier(text)
    item = self.ExtractItem(text)
    unit_price = self.ExtractUnitPrice(text)
    delivery_time = self.ExtractDeliveryTime(text)
    risk_notes = self.ExtractRiskNotes(text)
    return ExtractfromQoute(supplier,item,unit_price,delivery_time,risk_notes)


  def ExtractSupplier (self , text: str) -> str:
    match = re.search(r"Supplier\s+([A-Za-z\s&]+)",text)
    return match.group(1).strip() if match else "Unknown Supplier"

  def ExtractItem (self , text: str) -> str:
    match = re.search(r"Item\s+([A-Za-z\s&]+)",text)
    return match.group(1).strip() if match else "Unknown Item"

  def ExtractUnitPrice (self , text: str) -> str:
    match = re.search(r"Unit Price\s+([A-Za-z\s&]+)",text)
    return match.group(1).strip() if match else "Unknown Unit Price"

  def ExtractDeliveryTime (self , text: str) -> str:
    match = re.search(r"Delivery Time\s+([A-Za-z\s&]+)",text)
    return match.group(1).strip() if match else "Unknown Delivery Time"

  def ExtractRiskNotes (self , text: str) -> str:
    if "Internal Note" in text:
      return text.split("Internal Note")[1].strip()
    else:
      return "No Internal Note"
```

```python
import statistics

class EvaluatorAgent:

  def __init__(self, llm=None):
    self.llm = llm

  def evaluateOffer(self,query:str, offer: List[Dict[str,Any]]) -> Dict[str,Any]:

    if not offer :
      return {"recommendation":None, "reasoning":"No offers found"}

    prices = [o["unit price"] for o in offer if o.get("unit price", 0) > 0]
    avgprice = statistics.mean(prices) if prices else 1.0

    for o in offer:
      o["score"] = (
          avgprice / o["unit price"] if o.get("unit price", 0) > 0 else 0
      ) + (1/(1+o["delivery_days"])*10
        -(5 if "risk" in o.get("Risk_Notes"," ").lower() else 0))

    best_offer = max(offer, key=lambda x: x.get("score", 0))
    return {"recommendation": best_offer, "reasoning":"", "offers evaluated":offer}

  def generateReasoning(self, query :str, offer: List[Dict[str,Any]],best_offer:Dict[str,Any])->str:
    reasons = [f"{o['supplier']} (${o['unit price']}/unit, {o['delivery days']} days)" for o in offer]
    return ( f"comapared offers{','.join(reasons)}."
              f"best offer  {best_offer['supplier']}")

```
